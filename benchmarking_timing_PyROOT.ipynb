```python
#Imports - this cell is not timed because I was not sure how to count it.
#Python users pretty much always do imports so it didn't make sense to me to do.
import time
import pandas as pd
import numpy as np
import datetime as dt
#from R76Tools import *
import ROOT
import matplotlib.pyplot as plt
import glob
from IPython import display
```

```python
#Some functions to minimize programming time
frittspath = "/data/chocula/fritts/data/k100proc/midasrq/byseries/"
c = ROOT.TCanvas("c","my canvas",800,400)
c.Divide(2,1)
#c.cd(1)
#h1 = ROOT.TH2D("h1","",300,0,30e-6,300,0,10e6)
def benchmark(series):
    filepath=[]
    start = time.time()
    filepath = filepath+glob.glob(series+"/umn_*.root")
    eT = ROOT.TChain("rqDir/eventTree")
    zip1 = ROOT.TChain("rqDir/zip1")
    for file in filepath:
       eT.Add(file)
       zip1.Add(file)
    zip1.AddFriend(eT)
    #print(zip1.GetEntries())
    #c = ROOT.TCanvas("c","my canvas",800,1200)
    #c.Divide(2,1)
    c.cd(1)
    zip1.Draw("EventTime:PTOFamps")
    end = time.time()
    print(end - start)
    c.cd(2)
    zip1.Draw("PTOFamps")
    print(time.time() - start)
    print("(",time.time()-end,")")
   
```

```python
benchmark(frittspath+"07220706_2041")
c.SaveAs("hist1.png","png")
display.Image("./hist1.png")
```
